```python
#importing Libraries for missing value imputation

from fancyimpute import KNN
from fancyimpute import IterativeImputer
```

```python
#Handling catgorical values with get dummies

df_1=pd.get_dummies(df_1,drop_first=True)
```

```python
#Taking copies of the dataset for KNN & MICE

df_knn=df_1.copy()
df_mice=df_1.copy()
```

```python
#Assigning KNN

knn=KNN()

#Instantiating Knn

df_knn.iloc[:,:]=knn.fit_transform(df_knn)
```

```python
#Assigning mice

mice=IterativeImputer(random_state=102)

#Instantiating mice

df_mice.iloc[:,:]=mice.fit_transform(df_mice)
```

```python
#Reseting index to get correct index

df_knn.reset_index(inplace=True)
df_mice.reset_index(inplace=True)
```

```python
#dropping the index

df_knn.drop('index',axis=1,inplace=True)
df_mice.drop('index',axis=1,inplace=True)
```

```python
#Dummy encoding the categorical variables

df_knn_fin_1=pd.get_dummies(df_knn_fin_1,columns=['Car_segment', 'seller_type', 'owner_type', 'fuel_type',
       'transmission_type', 'no_of_seat'],drop_first=True)
```

```python
import statsmodels
import statsmodels.api as sm
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
#Train-Test split

x=df_knn_fin_1.drop(['selling_price'],axis=1)

y=df_knn_fin_1['selling_price']

#x=sm.add_constant(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20,test_size=0.20)
```

```python
df_knn_fin_1.shape
```

# LINEAR REGRESSION MODEL

```python
LR_model = LinearRegression()

LR_model_tr = LR_model.fit(x_train,y_train)

LR_model_ts = LR_model.fit(x_test,y_test)

print("Training score :",round(cross_val_score(LR_model_tr,x_train,y_train,cv=10).mean(),2))

print("Testing score :",round(cross_val_score(LR_model_ts,x_test,y_test,cv=10).mean(),2))

y_pred_tr = LR_model_tr.predict(x_train)

y_pred_ts = LR_model_ts.predict(x_test)

print("Training MAE :",round(mean_absolute_error(y_train,y_pred_tr),2))

print("Testing MAE :",round(mean_absolute_error(y_test,y_pred_ts),2))
```

# DECISION TREE MODEL

```python
DT_model = DecisionTreeRegressor(random_state=1)

DT_model_tr = DT_model.fit(x_train,y_train)

DT_model_ts = DT_model.fit(x_test,y_test)

print("Training score :",round(cross_val_score(DT_model_tr,x_train,y_train,cv=10).mean(),2))

print("Testing score :",round(cross_val_score(DT_model_ts,x_test,y_test,cv=10).mean(),2))

y_pred_tr = DT_model_tr.predict(x_train)

y_pred_ts = DT_model_ts.predict(x_test)

print("Training MAE :",round(mean_absolute_error(y_train,y_pred_tr),2))

print("Testing MAE :",round(mean_absolute_error(y_test,y_pred_ts),2))
```

# RANDOM FORREST MODEL 

```python
RF_model = RandomForestRegressor()

RF_model_tr = RF_model.fit(x_train,y_train)

RF_model_ts = RF_model.fit(x_test,y_test)

print("Training score :",round(cross_val_score(RF_model_tr,x_train,y_train,cv=10).mean(),2))

print("Testing score :",round(cross_val_score(RF_model_ts,x_test,y_test,cv=10).mean(),2))

y_pred_tr = RF_model_tr.predict(x_train)

y_pred_ts = RF_model_ts.predict(x_test)

print("Training MAE :",round(mean_absolute_error(y_train,y_pred_tr),2))

print("Testing MAE :",round(mean_absolute_error(y_test,y_pred_ts),2))
```
